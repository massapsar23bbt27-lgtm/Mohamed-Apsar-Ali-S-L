1. K-mer Analysis and Motif Searching:

```python
from collections import Counter

sequence = input("Enter the sequences:").upper()
kmer_len = int(input("enter the length:"))
target = input("Enter the target:").upper()
kmers_list = []

for i in range(len(sequence) - kmer_len + 1):
    kmers_list.append(sequence[i:i+kmer_len])

kmer_freq = Counter(kmers_list)
print("Total K-mers:", len(kmers_list))
print("Distinct K-mers:", len(set(kmers_list)))

for k, v in kmer_freq.items():
    if v > 1:
        print(f"'{k}' appears {v} times")

for idx in range(len(sequence) - len(target) + 1):
    if sequence[idx:idx+len(target)] == target:
        print(f"Target '{target}' located at index {idx}")
```

2. Grouping and Filtering Dictionaries

```python
from collections import defaultdict

records_count = int(input("Enter the records count:"))
mutations = []

for _ in range(records_count):
    gene_id = input("Enter the gene id:")
    mut_name = input("Enter the mutation name:")
    impact = input("Enter the impact:")
    freq = float(input("Enter the frequency:"))

    mutations.append({
        "gene": gene_id,
        "mutation": mut_name,
        "impact": impact,
        "frequency": freq
    })

critical_variants = defaultdict(list)

for mut in mutations:
    if mut["impact"].lower() == "high" and mut["frequency"] <= 0.05:
        critical_variants[mut["gene"]].append(mut["mutation"])

for g, v_list in critical_variants.items():
    print(f"Gene {g} has critical mutations: {', '.join(v_list)}")
```


3. Finding Shared Codons via Sets




```python
!pip install biopython
from Bio.Seq import Seq

strand_a = Seq(input("Enter the DNA strand A:").upper())
strand_b = Seq(input("Enter the DNA strand B:").upper())

codons_a = {strand_a[i:i+3] for i in range(0, len(strand_a) - 2, 3)}
codons_b = {strand_b[i:i+3] for i in range(0, len(strand_b) - 2, 3)}

common_codons = codons_a.intersection(codons_b)
print("Common Codons:", list(common_codons))

amino_acids = {str(c.translate()) for c in common_codons if len(c) == 3}
print("Common Amino Acids:", list(amino_acids))
```

4. Sequence Motifs and Consensus

```python
!pip install biopython
from Bio import motifs
from Bio.Seq import Seq

num_seqs = int(input("Enter the numbers:"))
dna_list = []

for _ in range(num_seqs):
    dna_list.append(Seq(input("Enter the DNA sequences:").upper()))

dna_motif = motifs.create(dna_list)

print("Position Frequency Matrix:")
print(dna_motif.counts)

print("Derived Consensus:", dna_motif.consensus)
```

5. Open Reading Frame Translation

```python
!pip install biopython
from Bio.Seq import Seq

nucleotides = Seq(input("Enter the DNA nucleotides: ").upper())

for offset in range(3):
    peptide = nucleotides[offset:].translate(to_stop=True)
    if peptide:
        print(f"Reading Frame {offset + 1} Peptide: {peptide}")
    else:
        print(f"Reading Frame {offset + 1} Peptide: [EMPTY]")
```
