1. DNA to Protein Translation

```python
dna_seq = input("Enter DNA sequence for translation: ").upper()
protein = ""
idx = 0

while idx < len(dna_seq) - 2:
    codon = dna_seq[idx:idx+3]

    if codon == "ATG":
        protein += "M"
    elif codon == "TTT" or codon == "TTC":
        protein += "F"
    elif codon in ["GCT", "GCC", "GCA", "GCG"]:
        protein += "A"
    elif codon in ["TAA", "TAG", "TGA"]:
        break
    else:
        protein += "X"

    idx += 3

print("Translated Protein:", protein)
```

2. DNA Palindrome (Reverse Complement) Checker

```python
dna_str = input("Enter DNA sequence to check for palindrome: ").upper()
is_palindrome = "YES"
length = len(dna_str)

for i in range(length // 2):
    front_base = dna_str[i]
    back_base = dna_str[length - 1 - i]

    comp = ""
    if front_base == "A": comp = "T"
    elif front_base == "T": comp = "A"
    elif front_base == "C": comp = "G"
    elif front_base == "G": comp = "C"

    if comp != back_base:
        is_palindrome = "NO"
        break

print("Is Palindromic:", is_palindrome)
```

3. GC Content and Sequence Identity Calculation

```python
seq1 = input("Enter first DNA sequence: ").upper()
seq2 = input("Enter second DNA sequence: ").upper()

gc_count = 0
match_count = 0
min_len = min(len(seq1), len(seq2))

for i in range(min_len):
    if seq1[i] == "G" or seq1[i] == "C":
        gc_count += 1
    if seq1[i] == seq2[i]:
        match_count += 1

gc_percent = (gc_count * 100) / min_len
identity_percent = (match_count * 100) / min_len

print("GC Percentage:", round(gc_percent, 2))
print("Identity Percentage:", round(identity_percent, 2))
```

4. DNA Molecular Weight Calculation

```python
dna_chain = input("Enter DNA sequence for weight calculation: ").upper()
total_weight = 0

for base in dna_chain:
    if base == "A":
        total_weight += 313.2
    elif base == "T":
        total_weight += 304.2
    elif base == "G":
        total_weight += 329.2
    elif base == "C":
        total_weight += 289.2

total_weight += 79.0

print("Total DNA Weight:", round(total_weight, 2))
```

5. Reverse Complement Generator

```python
forward_seq = input("Enter DNA sequence to get reverse complement: ").upper()
rev_comp = ""

for base in forward_seq:
    if base == "A":
        rev_comp = "T" + rev_comp
    elif base == "T":
        rev_comp = "A" + rev_comp
    elif base == "C":
        rev_comp = "G" + rev_comp
    elif base == "G":
        rev_comp = "C" + rev_comp

print("Reverse Complement:", rev_comp)
```
