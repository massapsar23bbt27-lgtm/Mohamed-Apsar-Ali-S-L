 1.RNA to Protein Translation:

```python
!pip install biopython
from Bio.Seq import Seq

rna_sequence = Seq(input("Enter the RNA sequences:").upper())
peptide = ""
idx = 0

while idx < len(rna_sequence) - 2:
    current_codon = rna_sequence[idx:idx+3]
    amino = current_codon.translate()

    if amino == "*":
        break

    peptide += str(amino)
    idx += 3

print(peptide)
```

2 Dot Plot Matrix Generator:

```python
!pip install biopython
from Bio.Seq import Seq

seq_a = Seq(input("Enter the sequence1:").upper())
seq_b = Seq(input("Enter the sequence2:").upper())

for x in range(len(seq_a)):
    row_str = ""
    for y in range(len(seq_b)):
        if seq_a[x] == seq_b[y]:
            row_str += "X "
        else:
            row_str += "- "
    print(row_str)
```

3.DNA Sequence Analysis - GC Content:

```python
!pip install biopython
from Bio.SeqUtils import gc_fraction

dna_chain = input("Enter the DNA sequences:").upper()
gc_val = gc_fraction(dna_chain) * 100

print(f"{gc_val:.2f}%")

if gc_val >= 60.0:
    print("High GC Content")
elif gc_val <= 40.0:
    print("Low GC Content")
else:
    print("Moderate GC Content")
```

4.ORF Finder / Translation by Reading Frames

```python
!pip install biopython
from Bio.Seq import Seq

raw_dna = Seq(input("Enter the DNA sequences:").upper())

for frame_idx in range(3):
    peptide_chain = ""

    for pos in range(frame_idx, len(raw_dna) - 2, 3):
        codon = raw_dna[pos:pos+3]
        aa = str(codon.translate())

        if aa == "*":
            break

        peptide_chain += aa

    if len(peptide_chain) > 0:
        print(f"Frame {frame_idx + 1}: {peptide_chain}")
    else:
        print(f"Frame {frame_idx + 1}: No translation")
```

5.Counting Nucleotides:

```python
!pip install biopython
from Bio.Seq import Seq

nuc_seq = Seq(input("Enter the DNA sequences:").upper())
freq_dict = {"A": 0, "C": 0, "G": 0, "T": 0, "U": 0}

for char in nuc_seq:
    if char in freq_dict:
        freq_dict[char] += 1

for base, count in freq_dict.items():
    if count > 0:
        print(f"Base {base}: {count}")
```
