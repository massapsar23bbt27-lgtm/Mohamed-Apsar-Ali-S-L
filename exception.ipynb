1. GC Content Calculation with Exception Handling

```python
user_sequence = input("Provide the DNA string: ").upper()

try:
    if not user_sequence:
        raise ValueError("The sequence provided is empty.")

    for nucleotide in user_sequence:
        if nucleotide not in "ACGT":
            raise ValueError(f"Unrecognized nucleotide: {nucleotide}")

    gc_total = 0
    for nucleotide in user_sequence:
        if nucleotide == "G" or nucleotide == "C":
            gc_total += 1

    gc_ratio = (gc_total / len(user_sequence)) * 100

    print("Provided Sequence:", user_sequence)
    print("Total GC Nucleotides:", gc_total)
    print("GC Ratio:", round(gc_ratio, 2), "%")

except ValueError as e:
    print("Input Error:", e)
```

2. DNA to Protein Mapper using Exception Handling

```python
translation_map = {
    "ATG": "M", "GCT": "A", "TGT": "C", "GGT": "G",
    "TAA": "*", "TAG": "*", "TGA": "*"
}

genetic_code = input("Enter a DNA sequence to translate: ").upper()

try:
    if len(genetic_code) == 0:
        raise ValueError("Sequence cannot be blank.")
    if len(genetic_code) % 3 != 0:
        raise ValueError("Sequence length must be perfectly divisible by 3.")

    polypeptide = ""

    for idx in range(0, len(genetic_code), 3):
        triplet = genetic_code[idx:idx+3]

        if triplet not in translation_map:
            raise KeyError(f"Codon {triplet} not found in the defined translation map.")

        amino_acid = translation_map[triplet]

        if amino_acid == "*":
            break

        polypeptide += amino_acid

    print("Genetic Code:", genetic_code)
    print("Polypeptide Chain:", polypeptide)

except ValueError as ve:
    print("Value Exception:", ve)
except KeyError as ke:
    print("Key Exception:", ke)
```

3. GC Content Calculation using Basic Iterations

```python
dna_chain = input("Input DNA for GC calculation: ").upper()
gc_amount = 0
total_length = 0

for char in dna_chain:
    total_length += 1

    if char == "G" or char == "C":
        gc_amount += 1

if total_length > 0:
    gc_fraction = (gc_amount / total_length) * 100
else:
    gc_fraction = 0

print("Sequence Analyzed :", dna_chain)
print("Sequence Length   :", total_length)
print("GC Nucleotides    :", gc_amount)
print("GC Percentage     :", round(gc_fraction, 2), "%")
```

4. Sequence Reversal using Iteration

```python
forward_strand = input("Enter a sequence to reverse: ").upper()
backward_strand = ""

for pos in range(len(forward_strand) - 1, -1, -1):
    backward_strand += forward_strand[pos]

print("Original Strand :", forward_strand)
print("Reversed Strand :", backward_strand)
```
